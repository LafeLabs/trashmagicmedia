```python
import json
f = open("data/sensordata.txt", "r")
textdata = f.read()
data = json.loads(textdata)
```

```python
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
pressurewave = []
temperaturewave = []
fwave = []
minwave = []
timewave = []
for x in data:
    #print(x['Pressure'])
    pressurewave.append(x['Pressure'])
    temperaturewave.append(x['Temperature'])
    fwave.append(32.0 + x['Temperature']*9/5)
    timewave.append(x['timestamp'])
    minwave.append((x['timestamp'] - data[0]['timestamp'])/60000)
plt.plot(minwave,fwave)
```

```python
plt.plot(minwave,pressurewave)
```

```python
plt.plot(fwave,pressurewave,"r+")
```

```python
fig = plt.figure(tight_layout=True)
ax = fig.add_subplot()
ax.plot(minwave,fwave)
ax.set_ylabel('Temperature[F]')
ax.set_xlabel('time[minutes]')
```

```python
fig = plt.figure(tight_layout=True)
ax = fig.add_subplot()
ax.plot(pressurewave,fwave,'r+')
ax.set_ylabel('Temperature[F]')
ax.set_xlabel('Pressure[hPa]')
```
